```python
class Node:
  def __init__(self,value):
    self.value= value
    self.next= None

#my linkedlist class
class LinkedList:
  def __init__(self):
    self.head= None
    self.tail= None
    self.size= 0


  def print_list(self):
    temp= self.head
    if temp is None:
      return "Linked List is empty "
    else:
      while temp is not None :
        print(temp.value, end="--->")
        temp= temp.next


  def prepend(self,value):
    new_node= Node(value)
    temp= self.head
    if temp is None:
      self.head = new_node
      self.tail= new_node
    else:
      self.head= new_node
      new_node.next= temp
    self.size+=1


  def append(self,value):
    new_node= Node(value)
    if self.head is None:
      self.head = new_node
      self.tail= new_node
    else:
      self.tail.next= new_node
      self.tail= new_node
    self.size+=1

  def pop_first(self):
    temp= self.head
    if temp is None:
      return None
    else:
      value= temp.value
      self.head= temp.next
      temp.next= None
      self.size-= 1
      return value


  def pop_last(self):
    temp= self.head
    if temp is None:
      return "None"
    else:
      while temp.next != self.tail:
       temp= temp.next
      temp.next= None
      value=self.tail.value
      self.tail= temp
      self.size-=1
      return value

  def set_value(self,index,value):
    temp= self.head
    if temp is None:
      return "None"
    elif index < 0 or index > self.size:
      return "None"
    else:
      for _ in range(index):
        temp = temp.next
      temp.value= value
      return temp.value==value


  def get_value(self,index):
    temp= self.head
    if temp is None:
      return "None"
    elif index < 0 or index > self.size:
      return "None"
    else:
      for _ in range(index):
        temp = temp.next
      return temp.value


  def insert(self,index, value):
    temp = self.head
    prev= None
    new_node= Node(value)
    if temp is None and index==0:
      self.prepend(value)
      return True
    elif index < 0 or index > self.size:
      return "None"
    else:
      for _ in range(index):
        prev= temp
        temp = temp.next
      prev.next= new_node
      new_node.next= temp
      self.size+=1


  def remove(self,index):
    temp= self.head
    prev= None
    if temp is None:
      return None
    elif index < 0 or index > self.size:
      return None
    elif index== 0:
      self.pop_first()
    elif index== self.size:
        self.pop_last()
    else:
      for _ in range(index):
        prev= temp
        temp= temp.next
      prev.next= temp.next
      temp.next = None
      self.size-=1
      return True


  def reverse(self):
    temp=self.head
    if temp is None:
      return None
    prev= None
    while temp.next != None :
      forward= temp.next
      temp.next= prev
      prev= temp
      temp= forward
    self.tail= self.head
    self.head= prev


  def find_middle(self):
    temp= self.head
    if self.head is None :
      return None
    back= temp
    forward= temp
    while forward is not None and forward.next is not None:
      back= back.next
      forward= forward.next.next

    return back.value


  def partition(self,value):

    # Initialize two new head nodes for the small and big lists
    small_head = small = Node(0)
    big_head = big = Node(0)

    # Iterate over the original list
    current = self.head
    while current:
        if current.value < value:
            small.next = current
            small = small.next
        else:
            big.next = current
            big = big.next
        current = current.next

    # At the end, ensure that the big list is terminated
    big.next = None

    # Connect the small list with the big list
    small.next = big_head.next

    # Reassign the head of the main list
    self.head = small_head.next

    # If there were no elements smaller than value, adjust head
    if small_head.next == big_head:
        self.head = big_head.next











ll= LinkedList()
ll.append(3)
ll.append(8)
ll.append(5)
ll.append(10)
ll.append(2)
ll.append(1)
ll.partition(5)
ll.print_list()


```

```python
array= [1,2,4,5,12]

def kth(array,k):
  if not array:
    return None

  if k > len(array):
    return None

  fast= 0
  slow= 0
  while fast < k:
    fast+=1
  while fast < len(array)-1:
    slow+=1
    fast+=1
  return array[slow]

print(kth(array,7))


```
